```python
import nltk
from nltk import word_tokenize, pos_tag, sent_tokenize
from nltk.chunk import RegexpParser

def extract_noun_phrases(paragraph):
    NP_List = []

    sentences = sent_tokenize(paragraph)

    grammar = "NP: {<DT>?<JJ>*<NN|NNS|NNP|NNPS>+}"
    chunk_parser = RegexpParser(grammar)

    for sentence in sentences:
        words = word_tokenize(sentence)

        tagged_words = pos_tag(words)
        tree = chunk_parser.parse(tagged_words)

        for subtree in tree.subtrees(filter=lambda t: t.label() == 'NP'):
            np = " ".join(word for word, tag in subtree.leaves())
            NP_List.append(np)

    return NP_List
```

```python
nltk.download('punkt_tab')

paragraph = "The quick brown fox jumps over the lazy dog. A smart AI system learns quickly."
nps = extract_noun_phrases(paragraph)

print("Noun Phrases:")
for np in nps:
    print(np)
```
