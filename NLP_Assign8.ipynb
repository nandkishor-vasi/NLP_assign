```python
df = pd.DataFrame(data, columns=["text", "Category"])

def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    return text

df['text'] = df['text'].apply(preprocess)

model = Pipeline([
    ("tfidf", TfidfVectorizer(
        stop_words="english",
        ngram_range=(1,2)
    )),
    ("clf", LinearSVC(class_weight="balanced"))
])

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
y_pred = cross_val_predict(model, df['text'], df['Category'], cv=skf)

print(classification_report(df['Category'], y_pred))

model.fit(df['text'], df['Category'])
```

```python
def predict(query):
    query = preprocess(query)
    return model.predict([query])[0]

print(predict("payment deducted but failed"))
print(predict("app crashes when I open it"))
print(predict("I cannot login to my account"))
print(predict("tell me about your app"))
```
