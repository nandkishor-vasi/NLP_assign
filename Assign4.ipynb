```python
from nltk.tokenize import word_tokenize
import nltk

nltk.download('punkt')
```

```python
nltk.download('punkt_tab')

corpus = [
    "The King is a strong man",
    "The Queen is a wise woman",
    "The Boy is a young man",
    "The Girl is a young woman",
    "The Prince is a young king",
    "The Princess is a young queen",
    "The Man is strong",
    "The Woman is wise",
]

tokenized_corpus = [word_tokenize(sentence.lower()) for sentence in corpus]
```

```python
!pip install gensim
from gensim.models import Word2Vec

cbow_model = Word2Vec(
    sentences=tokenized_corpus,
    vector_size=50,
    window=2,
    min_count=1,
    sg=0,
    workers=2,
    epochs=100,
)
```

```python
skip_gram_model = Word2Vec(
    sentences=tokenized_corpus,
    vector_size=50,
    window=2,
    min_count=1,
    sg=1,
    workers=2,
    epochs=200,
)
```

```python
cbow_model.wv.most_similar("king")
```

```python
print("CBOW model - Most similar to 'king':", cbow_model.wv.most_similar("king"))

print("Skip-gram model - Most similar to 'king':", skip_gram_model.wv.most_similar("king"))
```

```python
similarity = cbow_model.wv.similarity('king', 'queen')
print("Similarity between 'king' and 'queen' in CBOW model:", similarity)
similarity = skip_gram_model.wv.similarity('king', 'queen')
print("Similarity between 'king' and 'queen' in Skip-gram model:", similarity)
```

```python
print("CBOW analogy:")
print(cbow_model.wv.most_similar(
    positive=['king', 'woman'],
    negative=['man']
))

print("\nSkip-Gram analogy:")
print(skip_gram_model.wv.most_similar(
    positive=['king', 'woman'],
    negative=['man']
))
```
